### Soal 3

1. Gunakan database “Sakila”
2. Tampilkan field durasi rental, banyaknya film dalam setiap durasi rental, dan rata-rata durasi film dari table “film”. Kelompokkan semuanya berdasarkan durasi rentalnya
3. Karena rata-rata durasi film menghasilkan angka desimal, bulatkan hingga hanya 2 angka di belakang koma saja.


```python
sql_df(
    '''
    SELECT rental_duration as Durasi_Rental,
    COUNT(film_id) as Banyak_Film,
    ROUND(AVG(length), 2) as Rata_rata_Durasi_Film
    FROM film
    GROUP BY rental_duration
    ORDER BY rental_duration
    '''
)
```

### Soal 4

2. Dari table film, tampilkan title, durasi film, dan juga rating yang durasi filmnya lebih dari rata-rata durasi film total. 
3. Tampilkan 25 data yang diurutkan dari durasi terlama


```python
sql_df(
    '''
    SELECT title, length, rating
    FROM film
    WHERE length >
    (SELECT AVG(length) FROM film)
    ORDER BY length
    LIMIT 25
    '''
)
```

### Soal 6

1. Gunakan database “Sakila”
2. Tampilkan 15 daftar film yang memiliki huruf “K” pada akhir pada title. Tampilkan title, durasi, serta Bahasa pada film.
3. Sebagai catatan, lakukan join terlebih dahulu dari table “film” dan juga table “language”. 


```python
sql_df(
    '''
    SELECT film.title as Judul,
    film.length as Durasi,
    language.name as Bahasa_Film
    FROM film INNER JOIN language
    ON film.language_id = language.language_id
    WHERE film.title LIKE '%k'
    LIMIT 15;
    '''
)
```

### Soal 7

1. Gunakan database “Sakila”
2. Tampilkan Judul Film (Dari table “film”), First Name (Dari table “actor”), dan Last Name (Dari table “actor”) dari actor yang memiliki “actor_id” = 14
3. Sebagai catatan, lakukan join table terlebih dahulu antara table “film”, “film_actor”, dan table “actor”. 


```python
sql_df(
    '''
    SELECT film.title,
    actor.first_name,
    actor.last_name
    FROM film_actor 
    JOIN film 
    ON film_actor.film_id = film.film_id
    JOIN actor 
    ON actor.actor_id = film_actor.actor_id
    WHERE film_actor.actor_id = 14;
    '''
)
```

### Soal 9

1. Gunakan database “sakila”
2. Tampilkan nama genre (name) dari table “category” dan jumlah banyaknya film di setiap genrenya.
3. Lakukan join terlebih dahulu antara table “film”, “film_category”, dan “category”
4. Urutkan berdasarkan jumlah film di setiap kategorinya.


```python
sql_df(
    '''
    SELECT category.name as Genre,
    COUNT(film.title) as Banyak_Film
    FROM film
    LEFT JOIN film_category 
    ON film.film_id = film_category.film_id
    LEFT JOIN category 
    ON film_category.category_id = category.category_id
    GROUP BY category.name
    ORDER BY Banyak_Film;
    '''
)
```

### Soal 10

1. Gunakan database “Sakila”
2. Tampilkan title, description, length, serta rating dari table “film”. 
3. Tampilkan 10 judul yang memiliki judul dengan akhiran ‘h’ dan memiliki durasi lebih dari durasi rata-rata.
4. Urutkan berdasarkan judul nya.


```python
sql_df(
    '''
    SELECT title, description, length, rating
    FROM film
    WHERE title like '%h' AND
    length > (SELECT AVG(length) FROM film)
    ORDER BY title
    LIMIT 10;
    '''
)
```
